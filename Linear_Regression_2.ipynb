```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#number of data points
n=3
```

```python
#sample dataset
x = np.array([0.5,2.3,2.9])
y = np.array([1.4,1.9,3.2])
```

```python
#start with m=0 and c=0
m=0
c=0
learning_rate = 0.01
```

```python
for i in range(1,1000):
  y_predict =  m * x + c

  #calculate cost
  cost = (1/n * sum([value ** 2 for value in (y-y_predict)]))

  #plot after each intercation cost against m
  plt.scatter(c,cost)

  #calculate gradients
  dm = -(2) * sum(x * (y-y_predict))
  dc = -(2) * sum(y-y_predict)

  #update parameters
  m=m-learning_rate * dm
  c=c-learning_rate * dc

  print("m{}, c{}, cost{}, Iteration{}".format(m,c,cost,i))
plt.show()
```
